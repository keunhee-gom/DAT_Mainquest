```python
import random

class Character:
    def __init__(self, name, level, health, attack, defense):
        self.name = name
        self.level = level
        self.health = health
        self.attack = attack
        self.defense = defense

    
    def is_alive(self):  # 모든 인스턴스 메서드는 반드시 첫 번째 인자로 self를 받아야 함
        return self.health > 0
        
    def take_damage(self, damage):   # +damage
        if self.defense > damage:
            return            
        else:
            self.health -= (damage-self.defense)
            
    def attack_target(self, target):  # +target
        damage=random.randint(1, self.attack)
        print(f"{self.name}이(가) {target.name}에게 {damage}의 데미지를 입혔습니다!") # 잘 모르겠음
        target.take_damage(damage)
        print(f"\t{target.name}의 남은 체력: {target.health}\n")
        return damage # 이것의 의미는?
```

```python
class Player(Character):
    def __init__(self, name):
        super().__init__(name, 1, 100, 25, 5)  # 부모 클래스 초기화
        self.experience = 0  # 추가 속성
    
    def gain_experience(self,amount): # +amount
        self.experience+=amount
    
    def level_up(self):
        if self.experience>=50:
            self.level += 1
            self.attack += 10
            self.defense += 5
            self.experience = 0 # 레벨업 후 경험치 초기화
            print(f"\n♠{self.name} 레벨업♠ → 레벨: {self.level}, 공격력: {self.attack}, 방어력: {self.defense}")
        else:
            return # 이것의 의미
```

```python
class Monster(Character):
    def __init__(self, name,level):
        health = random.randint(10,30)*level # 여기선 왜 그냥 level일까 self.level이 아니라?
        attack = random.randint(5,20)*level
        defense = random.randint(1,5)*level
        super().__init__(name, level, health, attack, defense) 
        # 몬스터의 체력, 공격력, 방어력은 랜덤으로 먼저 계산하고,super()로 부모에게 있는 생성자 함수 사용
```

```python
def battle(player, monster):
    print()
    print('*'* 30)
    print(f"전투 시작!\n{player.name} vs {monster.name} (레벨 {monster.level}) ")
    print('*'* 30)
    
    while player.is_alive() and monster.is_alive():
        # 플레이어 공격
        player.attack_target(monster)
        if not monster.is_alive():
            player.gain_experience(monster.level*20)
            player.level_up()
            print()
            print('♥' * 14)
            print(" 전투 승리!")
            print('♥' * 14)
            break
         
        # 몬스터 공격
        monster.attack_target(player)
        if not player.is_alive():
            print()
            print('-' * 14)
            print("전투 패배")
            print('-' * 14)
            break
```

```python
def main():
    monster_dict ={'슬라임':1, '고블린':2,'오크':3}
    name= input('이름을 입력하세요: ')
    player=Player(name) # player=Player(input('이름을 입력하세요: '))
    # 변수로 name을 받아서 처리하는 방식이 더 읽기 좋고 유지보수에 유리
    
    for monster_name,monster_level in monster_dict.items(): # 딕셔너리의.items()를 사용하면 이름과 레벨을 동시애 가져옴
        monster=Monster(monster_name,monster_level)
        battle(player, monster)
        if not player.is_alive():
            print('-' * 14)
            print("게임오버")
            print('-' * 14)
            break
```

```python
main()
```
