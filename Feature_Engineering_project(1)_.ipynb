# **2. Z-Score Í≥ÑÏÇ∞**
## 1) Íµ¨Îß§Í∏àÏï°
Í∞Å Í≥†Í∞ùÏùò ÌèâÏÜå ÏÜåÎπÑ Ìå®ÌÑ¥ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Ïù¥ÏÉÅ Í±∞ÎûòÎ•º ÌÉêÏßÄ. Z-ScoreÍ∞Ä ÎÜíÏùÑÏàòÎ°ù ÌèâÏÜåÏôÄ Îã§Î•∏ Í±∞ÎûòÌå®ÌÑ¥ÏùÑ ÏùòÎØ∏

```python
# Í≥†Í∞ùÎ≥Ñ Íµ¨Îß§Í∏àÏï° ÌÜµÍ≥Ñ Í≥ÑÏÇ∞
# Ïã†Ïö©Ïπ¥Îìú Î≤àÌò∏(cc_num)Î≥ÑÎ°ú Î¨∂Í≥†, Í±∞Îûò Í∏àÏï°(amt) Ïª¨Îüº ÏÑ†ÌÉùÌï¥ÏÑú, ÌèâÍ∑†Í≥º ÌëúÏ§ÄÌé∏Ï∞® Í≥ÑÏÇ∞
amt_info = cc_df.groupby('cc_num')['amt'].agg(['mean', 'std'])
```

```python
amt_info.info()
```

```python
amt_info.reset_index()
```

```python
# amt_info Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ ÌîºÌÅ¥(pickle) ÌååÏùºÎ°ú Ï†ÄÏû•
# Ï†ÄÏû•Îêú ÌååÏùº Ïù¥Î¶ÑÏùÄ amt_info.pklÏù¥Í≥†, ÌòÑÏû¨ ÏûëÏóÖ ÎîîÎ†âÌÜ†Î¶¨Ïóê(./) Ï†ÄÏû•
amt_info.to_pickle('./amt_info.pkl')
```

```python
# Î©îÏù∏ Îç∞Ïù¥ÌÑ∞ÏôÄ ÌÜµÍ≥Ñ Ï†ïÎ≥¥ Î≥ëÌï©
cc_df = cc_df.merge(amt_info,on='cc_num',how='left')
```

```python
cc_df.head()
```

```python
# Z-Score Í≥ÑÏÇ∞
# Z-Score = (Í∞úÎ≥Ñ Í±∞ÎûòÍ∏àÏï° - Í≥†Í∞ù ÌèâÍ∑† Í±∞ÎûòÍ∏àÏï°) / Í≥†Í∞ù Í±∞ÎûòÍ∏àÏï° ÌëúÏ§ÄÌé∏Ï∞®
# ÌèâÍ∑†ÏóêÏÑú ÏñºÎßàÎÇò Îñ®Ïñ¥Ï†∏ ÏûàÎäîÏßÄÎ•º ÌëúÏ§ÄÌé∏Ï∞® Îã®ÏúÑÎ°ú ÎÇòÌÉÄÎÇ∏ Í∞í

cc_df['amt_z'] = (cc_df['amt']-cc_df['mean'])/cc_df['std']
```

```python
cc_df['amt_z'].describe()
```

```python
cc_df[cc_df['is_fraud']==1]
```

```python
print(cc_df.columns)
```

```python
# z-score Í≥ÑÏÇ∞Ïù¥ ÏôÑÎ£åÎêòÏóàÏúºÎãà mean, std Ïª¨ÎüºÏùÑ Ï†úÍ±∞
cc_df.drop(['mean','std'], axis = 1, inplace = True)
```

## 3) Í≤∞Ï†ú ÏãúÍ∞Ñ Feature ÏÉùÏÑ±
ÏÇ¨Í∏∞ Í±∞ÎûòÎäî ÌäπÏ†ï ÏãúÍ∞ÑÎåÄÏóê ÏßëÏ§ëÎêòÎäî Í≤ΩÌñ•. Í≥†Í∞ùÏùò ÌèâÏÜå Í±∞Îûò ÏãúÍ∞ÑÎåÄÏôÄ ÎπÑÍµêÌïòÏó¨ Ïù¥ÏÉÅ Ìå®ÌÑ¥ÏùÑ ÌÉêÏßÄ

```python
# trans_date_trans_time Ïª¨ÎüºÏóêÏÑú ÏãúÍ∞Ñ Í∞íÏùÑ Ï∂îÏ∂úÌïòÏó¨ hour Ïª¨ÎüºÏóê Ï†ÄÏû•
cc_df['trans_date_trans_time']=pd.to_datetime(cc_df['trans_date_trans_time'])
```

```python
cc_df['trans_date_trans_time']
```

```python
cc_df['hour']=cc_df['trans_date_trans_time'].dt.hour
```

```python
# Í≤∞Ï†úÏãúÍ∞ÑÏùÑ morning, afternoon, night, eveningÏúºÎ°ú Î∂ÑÎ•ò
def hour_func(x):
  if (x >= 6) & (x<12):
    return 'moring'
  elif (x >= 12) & (x < 18):
    return 'afternoon'
  elif (x >= 18) & (x < 23):
    return 'night'
  else:
    return 'evening'

cc_df['hour_cat']=cc_df['hour'].apply(hour_func)
```

```python
cc_df.head()
```

```python
time_dist = cc_df['hour_cat'].value_counts()
```

```python
time_dist
```

```python
# Í≥†Í∞ùÎ≥Ñ Ï†ÑÏ≤¥ Í±∞Îûò Ïàò
# Í∞Å Ïπ¥ÎìúÎ≤àÌò∏(cc_num)Î≥ÑÎ°ú Í∑∏Î£®Ìïë, Í∞Å Ïπ¥ÎìúÎ≤àÌò∏ Í∑∏Î£π ÎÇ¥ÏóêÏÑú amt Ïª¨ÎüºÏùò Í∞í Í∞úÏàò(Ï¶â, Í±∞Îûò ÌöüÏàò)Î•º ÏÖà
all_cnt = cc_df.groupby('cc_num')['amt'].count().reset_index()
```

```python
all_cnt.columns = ['cc_num', 'total_cnt']
```

```python
all_cnt
```

```python
# Í≥†Í∞ùÎ≥Ñ ÏãúÍ∞ÑÎåÄÎ≥Ñ Í±∞Îûò Ïàò
hour_cnt = cc_df.groupby(['cc_num', 'hour_cat'])['amt'].count().reset_index()
```

```python
hour_cnt
```

```python
hour_cnt.columns = ['cc_num', 'hour_cat', 'hour_cnt']
```

```python
 hour_cnt = hour_cnt.merge(all_cnt, on='cc_num', how='left')
```

```python
hour_cnt
```

```python
hour_cnt['hour_perc'] = hour_cnt['hour_cnt'] / hour_cnt['total_cnt']
```

```python
hour_cnt.head(10)
# Í≥†Í∞ù 1000Î≤àÏù¥ ÌèâÏÜå Ïò§Ï†Ñ(0.429)Ïóê Ï£ºÎ°ú Í±∞ÎûòÌïòÎäîÎç∞, Í∞ëÏûêÍ∏∞ ÏÉàÎ≤Ω(evening)Ïóê Í±∞ÎûòÌïúÎã§Î©¥ Ïù¥ÏÉÅ Ïã†Ìò∏Î°ú ÌåêÎã®
```

```python
hour_cnt.to_pickle('./hour_cnt.pkl')
```

```python
cc_df = cc_df.merge(hour_cnt,on=['cc_num','hour_cat'],how='left')
```

```python
cc_df.head(10)
```

```python
cc_df.drop(['trans_date_trans_time', 'hour', 'hour_cat'], axis =1 , inplace = True)
```

```python
# ÏúÑÎèÑÏôÄ Í≤ΩÎèÑ Í∞íÏúºÎ°ú Îëê ÏßÄÏ†ê ÏÇ¨Ïù¥Ïùò Í±∞Î¶¨Î•º Íµ¨ÌïòÍ∏∞
# Í≥†Í∞ùÏùò ÏúÑÏπòÏôÄ ÏÉÅÏ†êÏùò ÏúÑÏπò ÏÇ¨Ïù¥Ïùò Í±∞Î¶¨Î•º Í≥ÑÏÇ∞ÌïòÏó¨ distance Ïª¨ÎüºÏóê Ï†ÄÏû•
cc_df['distance'] = cc_df.apply(lambda x: distance((x['lat'], x['long']), (x['merch_lat'], x['merch_long'])).km, axis = 1)
```

```python
# Í±∞Î¶¨ Z-Score

dist_info = cc_df.groupby('cc_num')['distance'].agg(['mean', 'std']).reset_index()
```

```python
dist_info.to_pickle('./dist_info.pkl')
```

```python
cc_df=cc_df.merge(dist_info, on='cc_num', how='left')
```

```python
cc_df.shape
```

```python
cc_df['dist_z']= (cc_df['distance'] - cc_df['mean']) / cc_df['std']
```

```python
fraud_dist = cc_df.groupby('is_fraud')['dist_z'].describe()
```
